```python
import duckdb
import numpy as np

DB_PATH = "f:/Project/games/jt3/data/jt3.duckdb"
con = duckdb.connect(DB_PATH, read_only=True)
```

```python
queries = [
    "This element has atomic number 1",
    "The powerhouse of the cell",
]
```

```python
rows = con.execute(
    "SELECT clue_text, embedding FROM embeddings WHERE clue_text IN (SELECT unnest(?))",
    [queries],
).fetchall()

found = {text: np.array(emb, dtype=np.float32) for text, emb in rows}
missing = [q for q in queries if q not in found]

if missing:
    print(f"Not found ({len(missing)}): {missing}")

embeddings = [found[q] for q in queries if q in found]
print(f"Retrieved {len(embeddings)} embeddings, shape: {embeddings[0].shape if embeddings else 'N/A'}")
```

```python
con.close()
```
