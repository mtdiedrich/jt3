# Explore Embeddings

Query and inspect the embeddings table.

```python
import duckdb

conn = duckdb.connect("../data/jt3.duckdb", read_only=True)

%load_ext sql
%sql conn --alias duckdb
```

```python
%%sql
SELECT * FROM embeddings
```

```python
%%sql embeddings_frame <<
SELECT
    c.game_id,
    c.clue_id,
    c.text AS clue_text,
    c.correct_response AS answer_text,
    et.embedding AS clue_embedding,
    ea.embedding AS answer_embedding
FROM clues c
LEFT JOIN embeddings et
    ON  c.game_id = et.game_id
    AND c.round_index = et.round_index
    AND c.category_index = et.category_index
    AND c.clue_id = et.clue_id
    AND et.field = 'text'
LEFT JOIN embeddings ea
    ON  c.game_id = ea.game_id
    AND c.round_index = ea.round_index
    AND c.category_index = ea.category_index
    AND c.clue_id = ea.clue_id
    AND ea.field = 'correct_response'
# WHERE c.correct_response = 'Abraham Lincoln'
```

```python
embeddings_df = embeddings_frame.PolarsDataFrame()
embeddings_df
```

```python
game_id = 5811
clue_id = "J_1_1"
```

```python
%%sql closest_answers <<
WITH query AS (
    SELECT embedding
    FROM embeddings
    WHERE game_id = {{game_id}} AND clue_id = '{{clue_id}}' AND field = 'text'
    LIMIT 1
)
SELECT
    e.game_id,
    e.clue_id,
    e.source_text AS answer_text,
    list_cosine_similarity(q.embedding, e.embedding) AS similarity
FROM embeddings e, query q
WHERE e.field = 'correct_response'
ORDER BY similarity DESC
LIMIT 10
```

```python
%%sql clue_info <<
SELECT text AS clue_text, correct_response AS correct_answer
FROM clues
WHERE game_id = {{game_id}} AND clue_id = '{{clue_id}}'
LIMIT 1
```

```python
info = clue_info.PolarsDataFrame()
print(f"Clue:   {info['clue_text'][0]}")
print(f"Answer: {info['correct_answer'][0]}\n")

closest_answers_df = closest_answers.PolarsDataFrame()
closest_answers_df
```
